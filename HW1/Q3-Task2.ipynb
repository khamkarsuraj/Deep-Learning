```python
import tensorflow as tf
import numpy as np
import torch
import torchvision as tv
from torchvision import transforms, datasets
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
# Grab MNIST dataset
trainingSet = datasets.MNIST('', train=True, download=False, transform=transforms.Compose([transforms.ToTensor()]))
testingSet = datasets.MNIST('', train=False, download=False, transform=transforms.Compose([transforms.ToTensor()]))
train = torch.utils.data.DataLoader(trainingSet, batch_size=50, shuffle=True)
test = torch.utils.data.DataLoader(testingSet, batch_size=50, shuffle=True)
```

```python
def calculate_parameters(inputM1):
    val = sum(params.numel() for params in inputM1.parameters() if params.requires_grad)
    return val
```

```python
# Model Declarations
class M1 (nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784, 2)
        self.fc2 = nn.Linear(2, 5)
        self.fc3 = nn.Linear(5, 10)

    def forward(self, val):
        val = F.relu(self.fc1(val))
        val = F.relu(self.fc2(val))
        val = self.fc3(val)
        return val
    
# M 2 - 2 Hidden layer / 3270 Parameters
class M2 (nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784, 4)
        self.fc2 = nn.Linear(4, 8)
        self.fc3 = nn.Linear(8, 10)

    def forward(self, val):
        val = F.relu(self.fc1(val))
        val = F.relu(self.fc2(val))
        val = self.fc3(val)
        return val
    
# M 3 - 2 Hidden layer / 8280 Parameters
class M3 (nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784, 10)
        self.fc2 = nn.Linear(10, 20)
        self.fc3 = nn.Linear(20, 10)

    def forward(self, val):
        val = F.relu(self.fc1(val))
        val = F.relu(self.fc2(val))
        val = self.fc3(val)
        return val

# M 4 - 2 Hidden layer / 16640 Parameters
class M4 (nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784, 20)
        self.fc2 = nn.Linear(20, 30)
        self.fc3 = nn.Linear(30, 10)

    def forward(self, val):
        val = F.relu(self.fc1(val))
        val = F.relu(self.fc2(val))
        val = self.fc3(val)
        return val
    
# M 5 - 2 Hidden layer / 40785 Parameters
class M5 (nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784, 50)
        self.fc2 = nn.Linear(50, 25)
        self.fc3 = nn.Linear(25, 10)

    def forward(self, val):
        val = F.relu(self.fc1(val))
        val = F.relu(self.fc2(val))
        val = self.fc3(val)
        return val

# M 6 - 2 Hidden layer / 84060 Parameters
class M6 (nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784, 100)
        self.fc2 = nn.Linear(100, 50)
        self.fc3 = nn.Linear(50, 10)

    def forward(self, val):
        val = F.relu(self.fc1(val))
        val = F.relu(self.fc2(val))
        val = self.fc3(val)
        return val
    
# M 7 - 2 Hidden layer / 178110 Parameters
class M7 (nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784, 200)
        self.fc2 = nn.Linear(200, 100)
        self.fc3 = nn.Linear(100, 10)

    def forward(self, val):
        val = F.relu(self.fc1(val))
        val = F.relu(self.fc2(val))
        val = self.fc3(val)
        return val
    
# M 8 - 2 Hidden layer / 396210 Parameters
class M8 (nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784, 400)
        self.fc2 = nn.Linear(400, 200)
        self.fc3 = nn.Linear(200, 10)

    def forward(self, val):
        val = F.relu(self.fc1(val))
        val = F.relu(self.fc2(val))
        val = self.fc3(val)
        return val

# M 9 - 2 Hidden layer / 654310 Parameters
class M9 (nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784, 600)
        self.fc2 = nn.Linear(600, 300)
        self.fc3 = nn.Linear(300, 10)

    def forward(self, val):
        val = F.relu(self.fc1(val))
        val = F.relu(self.fc2(val))
        val = self.fc3(val)
        return val
    
# M 10 - 2 Hidden layer / 952410 Parameters
class M10 (nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784, 800)
        self.fc2 = nn.Linear(800, 400)
        self.fc3 = nn.Linear(400, 10)

    def forward(self, val):
        val = F.relu(self.fc1(val))
        val = F.relu(self.fc2(val))
        val = self.fc3(val)
        return val
```

```python
# Set up necessary auxilaries for neural net training
m1 = M1()
m2 = M2()
m3 = M3()
m4 = M4()
m5 = M5()
m6 = M6()
m7 = M7()
m8 = M8()
m9 = M9()
m10 = M10()
costFun = nn.CrossEntropyLoss()
m1Opt = optim.Adam(m1.parameters(), lr=0.001)
m2Opt = optim.Adam(m2.parameters(), lr=0.001)
m3Opt = optim.Adam(m3.parameters(), lr=0.001)
m4Opt = optim.Adam(m4.parameters(), lr=0.001)
m5Opt = optim.Adam(m5.parameters(), lr=0.001)
m6Opt = optim.Adam(m6.parameters(), lr=0.001)
m7Opt = optim.Adam(m7.parameters(), lr=0.001)
m8Opt = optim.Adam(m8.parameters(), lr=0.001)
m9Opt = optim.Adam(m9.parameters(), lr=0.001)
m10Opt = optim.Adam(m10.parameters(), lr=0.001)
m1Params = calculate_parameters(m1)
m2Params = calculate_parameters(m2)
m3Params = calculate_parameters(m3)
m4Params = calculate_parameters(m4)
m5Params = calculate_parameters(m5)
m6Params = calculate_parameters(m6)
m7Params = calculate_parameters(m7)
m8Params = calculate_parameters(m8)
m9Params = calculate_parameters(m9)
m10Params = calculate_parameters(m10)
```

```python
# Train all ms prior to testing and calc loss
EPOCHS = 70
c = 0
cList = []
for index in range(EPOCHS):
    cList.append(c)
    c += 1
    
    # M1 1
    for batch in train:
        inputImages, groundTruth = batch
        m1.zero_grad()
        output = m1(inputImages.view(-1,784))
        cost = costFun(output, groundTruth)
        cost.backward()
        m1Opt.step()

    # M1 2
    for batch in train:
        inputImages, groundTruth = batch
        m2.zero_grad()
        output = m2(inputImages.view(-1,784))
        cost = costFun(output, groundTruth)
        cost.backward()
        m2Opt.step()
        
    # M1 3
    for batch in train:
        inputImages, groundTruth = batch
        m3.zero_grad()
        output = m3(inputImages.view(-1,784))
        cost = costFun(output, groundTruth)
        cost.backward()
        m3Opt.step()
        
    # M1 4
    for batch in train:
        inputImages, groundTruth = batch
        m4.zero_grad()
        output = m4(inputImages.view(-1,784))
        cost = costFun(output, groundTruth)
        cost.backward()
        m4Opt.step()
            
    # M1 5
    for batch in train:
        inputImages, groundTruth = batch
        m5.zero_grad()
        output = m5(inputImages.view(-1,784))
        cost = costFun(output, groundTruth)
        cost.backward()
        m5Opt.step()
        
    # M1 6
    for batch in train:
        inputImages, groundTruth = batch
        m6.zero_grad()
        output = m6(inputImages.view(-1,784))
        cost = costFun(output, groundTruth)
        cost.backward()
        m6Opt.step()
        
    # M1 7
    for batch in train:
        inputImages, groundTruth = batch
        m7.zero_grad()
        output = m7(inputImages.view(-1,784))
        cost = costFun(output, groundTruth)
        cost.backward()
        m7Opt.step()
        
    # M1 8
    for batch in train:
        inputImages, groundTruth = batch
        m8.zero_grad()
        output = m8(inputImages.view(-1,784))
        cost = costFun(output, groundTruth)
        cost.backward()
        m8Opt.step()
        
    # M1 9
    for batch in train:
        inputImages, groundTruth = batch
        m9.zero_grad()
        output = m9(inputImages.view(-1,784))
        cost = costFun(output, groundTruth)
        cost.backward()
        m9Opt.step()
        
      # M1 10
    for batch in train:
        inputImages, groundTruth = batch
        m10.zero_grad()
        output = m10(inputImages.view(-1,784))
        cost = costFun(output, groundTruth)
        cost.backward()
        m10Opt.step()
```
